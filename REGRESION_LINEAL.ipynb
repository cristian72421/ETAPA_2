```python
#LIBRERIAS NECESARIAS

import matplotlib.pyplot as plt
from sklearn import linear_model
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
#CARGAR LA INFORMACION EN EL ARCHIVO CSV en la variable datos

datos = pd.read_csv("C:/Users/crist/Downloads/DATOS_REGRESION_LINEAL.csv" , sep= ",")
```

```python
datos
```

```python
#GRAFICA DE DISPERSION
datos.plot.scatter(x="metro", y="precio")
plt.show()
```

```python
regresion = linear_model.LinearRegression()
```

```python
#Agrego datos en una array o vector
metro =datos["metro"].values.reshape((-1,1))
```

```python
modelo = regresion.fit(metro, datos["precio"])
print("Interseccion (b)",modelo.intercept_)
print("Pendiente (m)", modelo.coef_)
```

```python
entrada = [[5],[15],[20],[25]]
predicciones = modelo.predict(entrada)
print(predicciones)
```

```python
datos.plot.scatter(x= "metro", y="precio", label='Datos originales')
plt.scatter(entrada, predicciones, color= 'yellow')
plt.plot(entrada, predicciones, color= 'black',label='linea de regresion')
plt.xlabel('metro')
plt.ylabel('precio')
plt.legend()
plt.show()
```
